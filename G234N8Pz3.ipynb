```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Администратор\\Downloads\\titanic.csv')
df
```

```python
survival_by_class = df.groupby('PClass')['Age'].sum()
plt.figure(figsize=(8, 6))
survival_by_class.plot(kind='bar', color='violet')
plt.xlabel('Класс пассажиров')
plt.ylabel('Возраст пассажиров')
plt.title('Зависимость возраста от класса пассажиров')
plt.xticks(rotation=0)
plt.show()
```
